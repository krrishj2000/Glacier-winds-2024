```python

# Leave one out cross validation

rmse_ours=[]
rmse_era5=[]
params=pd.read_csv('New_bigtable__old/NEWbigtableparams_backup_dontmes.csv')
params=params.iloc[0:15]
x_data=params[['station_z','lon','lat','Z_max']]
y_data=params['mean_temp_obs']
for i in range(15):
    
    from sklearn.linear_model import LinearRegression
    Lr=LinearRegression()
    leave=i
    model=Lr.fit(X=x_data.drop(index=leave),y=y_data.drop(index=leave))
    x_validate_sensitivity=x_data.iloc[i:i+1]
    y_validate_sensitivity=y_data.iloc[i:i+1]
    y_era5=params['mean_temp_era5'].iloc[i:i+1]
    y_predicted_sensitivity=model.predict(x_validate_sensitivity)
    rmse_ours.append(root_mean_squared_error(y_validate_sensitivity,y_predicted_sensitivity))
    rmse_era5.append(root_mean_squared_error(y_validate_sensitivity,y_era5))
    
results=pd.DataFrame()
results['RMSE']=np.zeros(30)
results['Estimator']=np.zeros(30)
results['RMSE'].iloc[0:15]=rmse_era5
results['RMSE'].iloc[15:30]=rmse_ours
results['Estimator'].iloc[0:15]='ERA5'
results['Estimator'].iloc[15:30]='Linear response'

plt.figure(figsize=(10,8))
plt.rcParams['font.size']=15
plt.xlabel("Mean Temperature")
sns.boxplot(data=results,x='Estimator',y='RMSE',color='gray')
```

## Hyperparameter tuning for sensitivity

```python
import itertools
import warnings; warnings.filterwarnings("ignore")
from sklearn.linear_model import HuberRegressor
def all_subsets(input_list):
    subsets = []
    for r in range(1,len(input_list) +1):
        subsets.extend(itertools.combinations(input_list, r))
    return subsets

# Example usage
my_list = ['glacier_width_at_station', 'Distance from centerline' , 'continentality', 'mean_wind_era5', 'T_range_continentality']
subsets = all_subsets(my_list)
params=pd.read_csv('New_bigtable/NEWbigtableparams_backup_dummy_newsensitivity.csv')
# params=params[pd.notna(params[list(subset)])]
params=params.iloc[0:15]
error_list=[]
ermin=[params['mean_wind_obs'].max()]
for index,subset in enumerate(subsets):
    rmse_ours=[]
    rmse_era5=[]
    params=pd.read_csv('New_bigtable/NEWbigtableparams_backup_dummy_newsensitivity.csv')
    params=params.iloc[0:15]
    params['Z_range']=params['Z_max']-params['Z_min']
    params['rel_z']=(params['station_z']-params['Z_min'])/(params['Z_max']-params['Z_min'])
    params['rel_distancefromterminus']=params['Distance from terminus']/params['glacier_length']
    params['rel_distancefromhead']=params['Distance from head']/params['glacier_length']
    params['rel_distancefromcenterline']=params['Distance from centerline']/params['glacier_width_at_station']
    params['local_aspect_normalised']=np.cos(np.degrees(params['aspect_1km_50p']))
    params['asrgi_aspect_normalised']=np.cos(np.degrees(params['rgi_aspect']))

    # x_data=params[['station_z','lon','glacier_length','channel_width_at_station','continentality']]
    x_data_sensitivity=params[list(subset)]
    y_data_sensitivity=params['mean_wind_obs']
    for i in range(15):
        
        from sklearn.linear_model import LinearRegression
        Lr=LinearRegression()
        leave=i
        model=Lr.fit(X=x_data_sensitivity.drop(index=leave),y=y_data_sensitivity.drop(index=leave))
        x_validate_sensitivity=x_data_sensitivity.iloc[i:i+1]
        y_validate_sensitivity=y_data_sensitivity.iloc[i:i+1]
        y_era5=[(params['mean_temp_era5'].max()-params['mean_temp_era5'].min())/(params['mean_wind_era5'].max()-params['mean_wind_era5'].min())]
        y_predicted_sensitivity=model.predict(x_validate_sensitivity)
        rmse_ours.append(root_mean_squared_error(y_validate_sensitivity,y_predicted_sensitivity))
        rmse_era5.append(root_mean_squared_error(y_validate_sensitivity,y_era5))
    error_list.append(np.mean(rmse_ours))
    # ermin=[np.median(rmse_ours) if u>np.median(rmse_ours) else u for u in ermin]
    print(np.argmin(error_list),subset,np.min(error_list),ermin)
    # print(subset)
```
